```python
import pandas as pd

df = pd.read_csv('./Datos/winemag-data-130k-v2.csv')
df.head()
```

Realice al menos 4 renombre de columna

```python


# Renombrar columnas
df = df.rename(columns={'country': 'pais', 'description': 'descripcion', 'points': 'puntos', 'price': 'precio'})
df.head(2)
```

```python
df_group  = df.groupby(['pais', 'designation']).aggregate({
    'precio': ['min', 'max', 'mean'], 'puntos': ['min', 'max', 'mean']
})

df_group
```

3 creación de nuevas columnas según criterio

```python

def etiquetar_pais_por_precio(row):
    if row['precio'] < 20:
        return 'Económico'
    elif 20 <= row['precio'] < 50:
        return 'Moderado'
    else:
        return 'Caro'

df['pais_por_precio'] = df.apply(etiquetar_pais_por_precio, axis=1)
df.head(2)
```

```python

def etiquetar_designacion_por_puntos(row):
    if row['puntos'] < 85:
        return 'Baja'
    elif 85 <= row['puntos'] < 90:
        return 'Media'
    else:
        return 'Alta'

df['designacion_por_puntos'] = df.apply(etiquetar_designacion_por_puntos, axis=1)
df.head(5)
```

```python

def etiquetar_variedad_por_precio(row):
    if row['precio'] < 20:
        return 'Variedad Económica'
    elif 20 <= row['precio'] < 50:
        return 'Variedad Moderada'
    else:
        return 'Variedad Cara'

df['variedad_por_precio'] = df.apply(etiquetar_variedad_por_precio, axis=1)
df.head(5)
```

Genere 4 reportes por agrupamiento de datos. Deberá elegir que reportes realizar

```python
#Reporte1: Promedio de puntos por país
p_puntos_pais = df.groupby('pais')['puntos'].mean().reset_index()
p_puntos_pais = p_puntos_pais.sort_values(by='puntos', ascending=False)

p_puntos_pais.head(7)
```

```python
# Reporte 2: Cantidad de vinos por país y designación 
c_vinos_pd = df.groupby(['pais', 'designation']).size().reset_index(name='cantidad_vinos')
c_vinos_pd = c_vinos_pd.sort_values(by='cantidad_vinos', ascending=False)

c_vinos_pd.head(6)
```

```python
# Reporte 3: Precio por variedad de vino 
precio_variedad = df.groupby('variety')['precio'].aggregate(['min', 'max', 'mean']).reset_index()
precio_variedad = precio_variedad.sort_values(by='max', ascending=False)

precio_variedad.head(6)
```

```python
# Reporte 4: Promedio de precio y puntos por variedad de vino 
prom_precio__variedad = df.groupby('variety').agg({'precio': 'mean', 'puntos': 'mean'}).reset_index()
prom_precio__variedad = prom_precio__variedad.sort_values(by='puntos', ascending=False)

prom_precio__variedad.head(6)
```

Al menos uno de estos datos agrupados deberán ser almacenados en excel o csv

```python
prom_precio__variedad.to_csv('prom_precio_variedad.csv', index=False)
```
